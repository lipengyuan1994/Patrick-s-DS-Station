```python
# å¯¼å…¥pandasï¼Œå¹¶ä¸”é‡å‘½åä¸ºpdã€‚
import pandas as pd

# é€šè¿‡äº’è”ç½‘è¯»å–æ³°å¦å°¼å…‹ä¹˜å®¢æ¡£æ¡ˆï¼Œå¹¶å­˜å‚¨åœ¨å˜é‡titanicä¸­ã€‚
titanic = pd.read_csv('http://biostat.mc.vanderbilt.edu/wiki/pub/Main/DataSets/titanic.txt')
```

```python
# äººå·¥é€‰å–pclassã€ageä»¥åŠsexä½œä¸ºåˆ¤åˆ«ä¹˜å®¢æ˜¯å¦èƒ½å¤Ÿç”Ÿè¿˜çš„ç‰¹å¾ã€‚
X = titanic[['pclass', 'age', 'sex']]
y = titanic['survived']
```

```python
# å¯¹äºç¼ºå¤±çš„å¹´é¾„ä¿¡æ¯ï¼Œæˆ‘ä»¬ä½¿ç”¨å…¨ä½“ä¹˜å®¢çš„å¹³å‡å¹´é¾„ä»£æ›¿ï¼Œè¿™æ ·å¯ä»¥åœ¨ä¿è¯é¡ºåˆ©è®­ç»ƒæ¨¡å‹çš„åŒæ—¶ï¼Œå°½å¯èƒ½ä¸å½±å“é¢„æµ‹ä»»åŠ¡ã€‚
X['age'] = X['age'].fillna(X['age'].mean())
```

```python
# å¯¹åŸå§‹æ•°æ®è¿›è¡Œåˆ†å‰²ï¼Œ25%çš„ä¹˜å®¢æ•°æ®ç”¨äºæµ‹è¯•ã€‚
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state = 33)
```

```python
# å¯¹ç±»åˆ«å‹ç‰¹å¾è¿›è¡Œè½¬åŒ–ï¼Œæˆä¸ºç‰¹å¾å‘é‡ã€‚
from sklearn.feature_extraction import DictVectorizer

vec = DictVectorizer(sparse=False)
X_train = vec.fit_transform(X_train.to_dict(orient='record'))
X_test = vec.transform(X_test.to_dict(orient='record'))
```

```python
# ä½¿ç”¨å•ä¸€å†³ç­–æ ‘è¿›è¡Œæ¨¡å‹è®­ç»ƒä»¥åŠé¢„æµ‹åˆ†æã€‚
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)
dtc_y_pred = dtc.predict(X_test)
```

```python
# ä½¿ç”¨éšæœºæ£®æ—åˆ†ç±»å™¨è¿›è¡Œé›†æˆæ¨¡å‹çš„è®­ç»ƒä»¥åŠé¢„æµ‹åˆ†æã€‚
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
rfc_y_pred = rfc.predict(X_test)
```

```python
# ä½¿ç”¨æ¢¯åº¦æå‡å†³ç­–æ ‘è¿›è¡Œé›†æˆæ¨¡å‹çš„è®­ç»ƒä»¥åŠé¢„æµ‹åˆ†æã€‚
from sklearn.ensemble import GradientBoostingClassifier

gbc = GradientBoostingClassifier()
gbc.fit(X_train, y_train)
gbc_y_pred = gbc.predict(X_test)
```

```python
# ä»sklearn.metricså¯¼å…¥classification_reportã€‚
from sklearn.metrics import classification_report

# è¾“å‡ºå•ä¸€å†³ç­–æ ‘åœ¨æµ‹è¯•é›†ä¸Šçš„åˆ†ç±»å‡†ç¡®æ€§ï¼Œä»¥åŠæ›´åŠ è¯¦ç»†çš„ç²¾ç¡®ç‡ã€å¬å›ç‡ã€F1æŒ‡æ ‡ã€‚
print ('The accuracy of decision tree is', dtc.score(X_test, y_test))
print (classification_report(dtc_y_pred, y_test))
```

```python
# è¾“å‡ºéšæœºæ£®æ—åˆ†ç±»å™¨åœ¨æµ‹è¯•é›†ä¸Šçš„åˆ†ç±»å‡†ç¡®æ€§ï¼Œä»¥åŠæ›´åŠ è¯¦ç»†çš„ç²¾ç¡®ç‡ã€å¬å›ç‡ã€F1æŒ‡æ ‡ã€‚
print ('The accuracy of random forest classifier is', rfc.score(X_test, y_test))
print (classification_report(rfc_y_pred, y_test))
```

```python
# è¾“å‡ºæ¢¯åº¦æå‡å†³ç­–æ ‘åœ¨æµ‹è¯•é›†ä¸Šçš„åˆ†ç±»å‡†ç¡®æ€§ï¼Œä»¥åŠæ›´åŠ è¯¦ç»†çš„ç²¾ç¡®ç‡ã€å¬å›ç‡ã€F1æŒ‡æ ‡ã€‚
print ('The accuracy of gradient tree boosting is', gbc.score(X_test, y_test))
print (classification_report(gbc_y_pred, y_test))
```

* é›†æˆæ¨¡å‹å¯ä»¥è¯´æ˜¯å®æˆ˜åº”ç”¨ä¸­æœ€ä¸ºå¸¸è§çš„ã€‚ç›¸æ¯”äºå…¶ä»–å•ä¸€çš„å­¦ä¹ æ¨¡å‹ï¼Œé›†æˆæ¨¡å‹å¯ä»¥æ•´åˆå¤šç§æ¨¡å‹ï¼Œæˆ–è€…å¤šæ¬¡å°±ä¸€ç§ç±»å‹çš„æ¨¡å‹è¿›è¡Œå»ºæ¨¡ã€‚ç”±äºæ¨¡å‹ä¼°è®¡å‚æ•°çš„è¿‡ç¨‹ä¹ŸåŒæ ·æ”¶åˆ°æ¦‚ç‡çš„å½±å“ï¼Œå…·æœ‰ä¸€å®šçš„ä¸ç¡®å®šæ€§ï¼›å› æ­¤ï¼Œé›†æˆæ¨¡å‹è™½ç„¶åœ¨è®­ç»ƒè¿‡ç¨‹ä¸­è¦æ¶ˆè€—æ›´å¤šçš„æ—¶é—´ï¼Œä½†æ˜¯å¾—åˆ°çš„ç»¼åˆæ¨¡å‹å¾€å¾€å…·æœ‰æ›´é«˜çš„è¡¨ç°æ€§èƒ½å’Œæ›´å¥½çš„ç¨³å®šæ€§ã€‚
* The ensemble model is arguably the most common in real-world applications. Compared to other single learning models, the ensemble model can integrate multiple models or integrate the same single model many times. Since the process of estimating parameters of the model is also affected by the probability, it has certain uncertainty; therefore, although the ensemble model consumes more time in the training process, the obtained comprehensive model tends to have higher performance and better stability.
