```python
# å¯¼å…¥pandasä¸numpyå·¥å…·åŒ…ã€‚
import pandas as pd
import numpy as np

# åˆ›å»ºç‰¹å¾åˆ—è¡¨ã€‚
column_names = ['Sample code number', 'Clump Thickness', 'Uniformity of Cell Size', 'Uniformity of Cell Shape', 'Marginal Adhesion', 'Single Epithelial Cell Size', 'Bare Nuclei', 'Bland Chromatin', 'Normal Nucleoli', 'Mitoses', 'Class']

# ä½¿ç”¨pandas.read_csvå‡½æ•°ä»äº’è”ç½‘è¯»å–æŒ‡å®šæ•°æ®ã€‚
data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/breast-cancer-wisconsin.data', names = column_names )

# å°†?æ›¿æ¢ä¸ºæ ‡å‡†ç¼ºå¤±å€¼è¡¨ç¤ºã€‚
data = data.replace(to_replace='?', value=np.nan)
# ä¸¢å¼ƒå¸¦æœ‰ç¼ºå¤±å€¼çš„æ•°æ®ï¼ˆåªè¦æœ‰ä¸€ä¸ªç»´åº¦æœ‰ç¼ºå¤±ï¼‰ã€‚
data = data.dropna(how='any')

# è¾“å‡ºdataçš„æ•°æ®é‡å’Œç»´åº¦ã€‚
data.shape
```

```python
# use sklearn.model_selection to split train and test data
from sklearn.model_selection import train_test_split

# éšæœºé‡‡æ ·25%çš„æ•°æ®ç”¨äºæµ‹è¯•ï¼Œå‰©ä¸‹çš„75%ç”¨äºæ„å»ºè®­ç»ƒé›†åˆã€‚
X_train, X_test, y_train, y_test = train_test_split(data[column_names[1:10]], data[column_names[10]], test_size=0.25, random_state=33)
```

```python
# æŸ¥éªŒè®­ç»ƒæ ·æœ¬çš„æ•°é‡å’Œç±»åˆ«åˆ†å¸ƒã€‚
y_train.value_counts()
```

```python
# æŸ¥éªŒæµ‹è¯•æ ·æœ¬çš„æ•°é‡å’Œç±»åˆ«åˆ†å¸ƒã€‚
y_test.value_counts()
```

```python
# ä»sklearn.preprocessingé‡Œå¯¼å…¥StandardScalerã€‚
from sklearn.preprocessing import StandardScaler
# ä»sklearn.linear_modelé‡Œå¯¼å…¥LogisticRegressionä¸SGDClassifierã€‚
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import SGDClassifier

# æ ‡å‡†åŒ–æ•°æ®ï¼Œä¿è¯æ¯ä¸ªç»´åº¦çš„ç‰¹å¾æ•°æ®æ–¹å·®ä¸º1ï¼Œå‡å€¼ä¸º0ã€‚ä½¿å¾—é¢„æµ‹ç»“æœä¸ä¼šè¢«æŸäº›ç»´åº¦è¿‡å¤§çš„ç‰¹å¾å€¼è€Œä¸»å¯¼ã€‚
ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)
```

```python
# initialize LogisticRegression and SGDClassifierã€‚
lr = LogisticRegression()
sgdc = SGDClassifier()

# è°ƒç”¨LogisticRegressionä¸­çš„fitå‡½æ•°/æ¨¡å—ç”¨æ¥è®­ç»ƒæ¨¡å‹å‚æ•°ã€‚
lr.fit(X_train, y_train)
# ä½¿ç”¨è®­ç»ƒå¥½çš„æ¨¡å‹lrå¯¹X_testè¿›è¡Œé¢„æµ‹ï¼Œç»“æœå‚¨å­˜åœ¨å˜é‡lr_y_predictä¸­ã€‚
lr_y_predict = lr.predict(X_test)

# è°ƒç”¨SGDClassifierä¸­çš„fitå‡½æ•°/æ¨¡å—ç”¨æ¥è®­ç»ƒæ¨¡å‹å‚æ•°ã€‚
sgdc.fit(X_train, y_train)
# ä½¿ç”¨è®­ç»ƒå¥½çš„æ¨¡å‹sgdcå¯¹X_testè¿›è¡Œé¢„æµ‹ï¼Œç»“æœå‚¨å­˜åœ¨å˜é‡sgdc_y_predictä¸­ã€‚
sgdc_y_predict = sgdc.predict(X_test)
```

```python
# ä»sklearn.metricsé‡Œå¯¼å…¥classification_reportæ¨¡å—ã€‚
from sklearn.metrics import classification_report

# ä½¿ç”¨é€»è¾‘æ–¯è’‚å›å½’æ¨¡å‹è‡ªå¸¦çš„è¯„åˆ†å‡½æ•°scoreè·å¾—æ¨¡å‹åœ¨æµ‹è¯•é›†ä¸Šçš„å‡†ç¡®æ€§ç»“æœã€‚
print ('Accuracy of LR Classifier:', lr.score(X_test, y_test))
# åˆ©ç”¨classification_reportæ¨¡å—è·å¾—LogisticRegressionå…¶ä»–ä¸‰ä¸ªæŒ‡æ ‡çš„ç»“æœã€‚
print (classification_report(y_test, lr_y_predict, target_names=['Benign', 'Malignant']))
```

```python
 # ä½¿ç”¨éšæœºæ¢¯åº¦ä¸‹é™æ¨¡å‹è‡ªå¸¦çš„è¯„åˆ†å‡½æ•°scoreè·å¾—æ¨¡å‹åœ¨æµ‹è¯•é›†ä¸Šçš„å‡†ç¡®æ€§ç»“æœã€‚
print ('Accuarcy of SGD Classifier:', sgdc.score(X_test, y_test))
# åˆ©ç”¨classification_reportæ¨¡å—è·å¾—SGDClassifierå…¶ä»–ä¸‰ä¸ªæŒ‡æ ‡çš„ç»“æœã€‚
print (classification_report(y_test, sgdc_y_predict, target_names=['Benign', 'Malignant']))
```

From the two reports above we can find that the LogisticRegression has a higher accuracy score than SGDClassifier
