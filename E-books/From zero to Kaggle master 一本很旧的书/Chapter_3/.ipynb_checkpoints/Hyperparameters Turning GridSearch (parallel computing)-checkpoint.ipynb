```python
# ä»sklearn.datasetsä¸­å¯¼å…¥20ç±»æ–°é—»æ–‡æœ¬æŠ“å–å™¨ã€‚
from sklearn.datasets import fetch_20newsgroups
# å¯¼å…¥numpyï¼Œå¹¶ä¸”é‡å‘½åä¸ºnpã€‚
import numpy as np

# ä½¿ç”¨æ–°é—»æŠ“å–å™¨ä»äº’è”ç½‘ä¸Šä¸‹è½½æ‰€æœ‰æ•°æ®ï¼Œå¹¶ä¸”å­˜å‚¨åœ¨å˜é‡newsä¸­ã€‚
news = fetch_20newsgroups(subset='all')

from sklearn.model_selection import train_test_split

# å¯¹å‰3000æ¡æ–°é—»æ–‡æœ¬è¿›è¡Œæ•°æ®åˆ†å‰²ï¼Œ25%æ–‡æœ¬ç”¨äºæœªæ¥æµ‹è¯•ã€‚
X_train, X_test, y_train, y_test = train_test_split(news.data[:3000], news.target[:3000], test_size=0.25, random_state=33)

# å¯¼å…¥æ”¯æŒå‘é‡æœºï¼ˆåˆ†ç±»ï¼‰æ¨¡å‹ã€‚
from sklearn.svm import SVC
# å¯¼å…¥TfidfVectorizeræ–‡æœ¬æŠ½å–å™¨ã€‚
from sklearn.feature_extraction.text import TfidfVectorizer
# å¯¼å…¥Pipelineã€‚
from sklearn.pipeline import Pipeline

#ä½¿ç”¨Pipeline ç®€åŒ–ç³»ç»Ÿæ­å»ºæµç¨‹ï¼Œå°†æ–‡æœ¬æŠ½å–ä¸åˆ†ç±»å™¨æ¨¡å‹ä¸²è”èµ·æ¥ã€‚
clf = Pipeline([('vect', TfidfVectorizer(stop_words='english', analyzer='word')), ('svc', SVC())])

# è¿™é‡Œéœ€è¦è¯•éªŒçš„2ä¸ªè¶…å‚æ•°çš„çš„ä¸ªæ•°åˆ†åˆ«æ˜¯4ã€3ï¼Œ svc__gammaçš„å‚æ•°å…±æœ‰10^-2, 10^-1... ã€‚è¿™æ ·æˆ‘ä»¬ä¸€å…±æœ‰12ç§çš„è¶…å‚æ•°ç»„åˆï¼Œ12ä¸ªä¸åŒå‚æ•°ä¸‹çš„æ¨¡å‹ã€‚
parameters = {'svc__gamma': np.logspace(-2, 1, 4), 'svc__C': np.logspace(-1, 1, 3)}

# ä»sklearn.model_selectionä¸­å¯¼å…¥ç½‘æ ¼æœç´¢æ¨¡å—GridSearchCVã€‚
from sklearn.model_selection import GridSearchCV

# å°†12ç»„å‚æ•°ç»„åˆä»¥åŠåˆå§‹åŒ–çš„PiplineåŒ…æ‹¬3æŠ˜äº¤å‰éªŒè¯çš„è¦æ±‚å…¨éƒ¨å‘ŠçŸ¥GridSearchCVã€‚è¯·å¤§å®¶åŠ¡å¿…æ³¨æ„refit=Trueè¿™æ ·ä¸€ä¸ªè®¾å®š ã€‚
gs = GridSearchCV(clf, parameters, verbose=2, refit=True, cv=3)


# æ‰§è¡Œå•çº¿ç¨‹ç½‘æ ¼æœç´¢ã€‚
%time _= gs.fit(X_train, y_train)
gs.best_params_, gs.best_score_

# è¾“å‡ºæœ€ä½³æ¨¡å‹åœ¨æµ‹è¯•é›†ä¸Šçš„å‡†ç¡®æ€§ã€‚
print (gs.score(X_test, y_test))
```

# Parallel GridSearch
* using multicore processor to do parallel gridsearch

```python
# ä»sklearn.datasetsä¸­å¯¼å…¥20ç±»æ–°é—»æ–‡æœ¬æŠ“å–å™¨ã€‚
from sklearn.datasets import fetch_20newsgroups
# å¯¼å…¥numpyï¼Œå¹¶ä¸”é‡å‘½åä¸ºnpã€‚
import numpy as np

# ä½¿ç”¨æ–°é—»æŠ“å–å™¨ä»äº’è”ç½‘ä¸Šä¸‹è½½æ‰€æœ‰æ•°æ®ï¼Œå¹¶ä¸”å­˜å‚¨åœ¨å˜é‡newsä¸­ã€‚
news = fetch_20newsgroups(subset='all')

from sklearn.model_selection import train_test_split

# å¯¹å‰3000æ¡æ–°é—»æ–‡æœ¬è¿›è¡Œæ•°æ®åˆ†å‰²ï¼Œ25%æ–‡æœ¬ç”¨äºæœªæ¥æµ‹è¯•ã€‚
X_train, X_test, y_train, y_test = train_test_split(news.data[:3000], news.target[:3000], test_size=0.25, random_state=33)

# å¯¼å…¥æ”¯æŒå‘é‡æœºï¼ˆåˆ†ç±»ï¼‰æ¨¡å‹ã€‚
from sklearn.svm import SVC

# å¯¼å…¥TfidfVectorizeræ–‡æœ¬æŠ½å–å™¨ã€‚
from sklearn.feature_extraction.text import TfidfVectorizer
# å¯¼å…¥Pipelineã€‚
from sklearn.pipeline import Pipeline

#ä½¿ç”¨Pipeline ç®€åŒ–ç³»ç»Ÿæ­å»ºæµç¨‹ï¼Œå°†æ–‡æœ¬æŠ½å–ä¸åˆ†ç±»å™¨æ¨¡å‹ä¸²è”èµ·æ¥ã€‚
clf = Pipeline([('vect', TfidfVectorizer(stop_words='english', analyzer='word')), ('svc', SVC())])

# è¿™é‡Œéœ€è¦è¯•éªŒçš„2ä¸ªè¶…å‚æ•°çš„çš„ä¸ªæ•°åˆ†åˆ«æ˜¯4ã€3ï¼Œ svc__gammaçš„å‚æ•°å…±æœ‰10^-2, 10^-1... ã€‚è¿™æ ·æˆ‘ä»¬ä¸€å…±æœ‰12ç§çš„è¶…å‚æ•°ç»„åˆï¼Œ12ä¸ªä¸åŒå‚æ•°ä¸‹çš„æ¨¡å‹ã€‚
parameters = {'svc__gamma': np.logspace(-2, 1, 4), 'svc__C': np.logspace(-1, 1, 3)}

# ä»sklearn.model_selectionä¸­å¯¼å…¥ç½‘æ ¼æœç´¢æ¨¡å—GridSearchCVã€‚
from sklearn.model_selection import GridSearchCV

# initialize GridSearchï¼Œn_jobs=-1 represents to use all CPUs on this machineã€‚
gs = GridSearchCV(clf, parameters, verbose=2, refit=True, cv=3, n_jobs=-1)

# æ‰§è¡Œå¤šçº¿ç¨‹å¹¶è¡Œç½‘æ ¼æœç´¢ã€‚
%time _= gs.fit(X_train, y_train)
gs.best_params_, gs.best_score_

# è¾“å‡ºæœ€ä½³æ¨¡å‹åœ¨æµ‹è¯•é›†ä¸Šçš„å‡†ç¡®æ€§ã€‚
print (gs.score(X_test, y_test))
```
