```python
# ä»sklearn.datasets å¯¼å…¥ irisæ•°æ®åŠ è½½å™¨ã€‚
from sklearn.datasets import load_iris
# ä½¿ç”¨åŠ è½½å™¨è¯»å–æ•°æ®å¹¶ä¸”å­˜å…¥å˜é‡irisã€‚
iris = load_iris()
# æŸ¥éªŒæ•°æ®è§„æ¨¡ã€‚
iris.data.shape
```

```python
# æŸ¥çœ‹æ•°æ®è¯´æ˜ã€‚å¯¹äºä¸€åæœºå™¨å­¦ä¹ çš„å®è·µè€…æ¥è®²ï¼Œè¿™æ˜¯ä¸€ä¸ªå¥½ä¹ æƒ¯ã€‚
# View the data description. This is a good habit for a machine learning practitioner.
print (iris.DESCR)
```

```python

from sklearn.model_selection import train_test_split
# ä»ä½¿ç”¨train_test_splitï¼Œåˆ©ç”¨éšæœºç§å­random_stateé‡‡æ ·25%çš„æ•°æ®ä½œä¸ºæµ‹è¯•é›†ã€‚
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.25, random_state=33)
```

```python
# ä»sklearn.preprocessingé‡Œé€‰æ‹©å¯¼å…¥æ•°æ®æ ‡å‡†åŒ–æ¨¡å—ã€‚
from sklearn.preprocessing import StandardScaler
# ä»sklearn.neighborsé‡Œé€‰æ‹©å¯¼å…¥KNeighborsClassifierï¼Œå³Kè¿‘é‚»åˆ†ç±»å™¨ã€‚
from sklearn.neighbors import KNeighborsClassifier

# å¯¹è®­ç»ƒå’Œæµ‹è¯•çš„ç‰¹å¾æ•°æ®è¿›è¡Œæ ‡å‡†åŒ–ã€‚
ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)

# ä½¿ç”¨Kè¿‘é‚»åˆ†ç±»å™¨å¯¹æµ‹è¯•æ•°æ®è¿›è¡Œç±»åˆ«é¢„æµ‹ï¼Œé¢„æµ‹ç»“æœå‚¨å­˜åœ¨å˜é‡y_predictä¸­ã€‚
knc = KNeighborsClassifier()
knc.fit(X_train, y_train)
y_predict = knc.predict(X_test)
```

```python
# ä½¿ç”¨æ¨¡å‹è‡ªå¸¦çš„è¯„ä¼°å‡½æ•°è¿›è¡Œå‡†ç¡®æ€§æµ‹è¯„ã€‚
print ('The accuracy of K-Nearest Neighbor Classifier is', knc.score(X_test, y_test) )
```

```python
# ä¾ç„¶ä½¿ç”¨sklearn.metricsé‡Œé¢çš„classification_reportæ¨¡å—å¯¹é¢„æµ‹ç»“æœåšæ›´åŠ è¯¦ç»†çš„åˆ†æã€‚
from sklearn.metrics import classification_report
print (classification_report(y_test, y_predict, target_names=iris.target_names))
```

* K ä¸´è¿‘ï¼ˆåˆ†ç±»ï¼‰æ˜¯éå¸¸ç›´è§‚çš„æœºå™¨å­¦ä¹ æ¨¡å‹ï¼Œå› æ­¤æ·±å—å¹¿å¤§åˆå­¦è€…çš„å–œçˆ±ï¼Œè®¸å¤šæ•™ç§‘ä¹¦å¸¸å¸¸ä»¥æ­¤æ¨¡å‹ä¸ºä¾‹æŠ›ç –å¼•ç‰ï¼Œä¾¿è¶³ä»¥çœ‹å‡ºå…¶ä¸ä»…ç‰¹åˆ«ï¼Œè€Œä¸”å°šæœ‰ç‘•ç–µä¹‹å¤„ã€‚ç»†å¿ƒçš„è¯»è€…ä¼šå‘ç°ï¼ŒKä¸´è¿‘ç®—æ³•ä¸å…¶ä»–æ¨¡å‹æœ€å¤§çš„ä¸åŒåœ¨äºï¼š è¯¥æ¨¡å‹æ²¡æœ‰å‚æ•°è®­ç»ƒè¿‡ç¨‹ï¼Œè€Œåªæ˜¯æ ¹æ®æµ‹è¯•æ ·æœ¬åœ¨è®­ç»ƒæ•°æ®çš„åˆ†å¸ƒç›´æ¥ä½œå‡ºåˆ†ç±»å†³ç­–ã€‚å› æ­¤ï¼ŒKä¸´è¿‘å±äºæ— å‚æ•°æ¨¡å‹ Nonparametric model ä¸­çš„ä¸€ç§ã€‚ç„¶è€Œï¼Œæ­£æ˜¯è¿™æ ·çš„å†³ç­–ç®—æ³•ï¼Œå¯¼è‡´äº†å…¶éå¸¸é«˜çš„è®¡ç®—å¤æ‚åº¦å’Œå†…å­˜æ¶ˆè€—ã€‚å› ä¸ºè¯¥æ¨¡å‹æ¯å¤„ç†ä¸€ä¸ªæµ‹è¯•æ ·æœ¬ï¼Œéƒ½éœ€è¦å¯¹æ‰€æœ‰é¢„å…ˆåŠ è½½åœ¨å†…å­˜çš„è®­ç»ƒæ ·æœ¬è¿›è¡Œéå†ï¼Œé€ä¸€è®¡ç®—ç›¸ä¼¼åº¦ã€æ’åºå¹¶ä¸”é€‰å–Kä¸ªæœ€ä¸´è¿‘è®­ç»ƒæ ·æœ¬çš„æ ‡è®°ï¼Œè¿›è€Œåšå‡ºåˆ†ç±»å†³ç­–ã€‚è¿™æ˜¯å¹³æ–¹çº§åˆ«çš„ç®—æ³•å¤æ‚åº¦ï¼Œä¸€æ—¦æ•°æ®è§„æ¨¡ç¨å¤§ï¼Œä½¿ç”¨è€…ä¾¿éœ€è¦æƒè¡¡æ›´å¤šçš„è®¡ç®—æ—¶é—´çš„ä»£ä»·ã€‚
* K-Nearest Neighbor(classifier) is a very intuitive machine learning model, so it is very popular among beginners. Many textbooks often use this model as an example to show that it is not only special, but also flawed. Careful readers will find that the biggest difference between the K-Nearest Neighbor algorithm and other models is that the model has no parameter training process, but only makes the classification decision directly based on the distribution of the training data. Therefore, K-Nearest Neighbor is one of the nonparametric models. However, it is such a decision algorithm that leads to its very high computational complexity and memory consumption. Because each time a model is processed, it needs to traverse all the training samples preloaded in memory, calculate the similarity, sort and select the markers of the K nearest training samples one by one, and then make the classification decision. This is the algorithmic complexity of the square level. Once the data size is slightly larger, the user needs to consider the cost of more computation time.
