```python
# å¯¼å…¥pandasç”¨äºæ•°æ®åˆ†æã€‚
import pandas as pd
# åˆ©ç”¨pandasçš„read_csvæ¨¡å—ç›´æ¥ä»äº’è”ç½‘æ”¶é›†æ³°å¦å°¼å…‹å·ä¹˜å®¢æ•°æ®ã€‚
titanic = pd.read_csv('http://biostat.mc.vanderbilt.edu/wiki/pub/Main/DataSets/titanic.txt')
```

```python
# è§‚å¯Ÿä¸€ä¸‹å‰å‡ è¡Œæ•°æ®ï¼Œå¯ä»¥å‘ç°ï¼Œæ•°æ®ç§ç±»å„å¼‚ï¼Œæ•°å€¼å‹ã€ç±»åˆ«å‹ï¼Œç”šè‡³è¿˜æœ‰ç¼ºå¤±æ•°æ®ã€‚
titanic.head()
```

```python
# ä½¿ç”¨pandasï¼Œæ•°æ®éƒ½è½¬å…¥pandasç‹¬æœ‰çš„dataframeæ ¼å¼ï¼ˆäºŒç»´æ•°æ®è¡¨æ ¼ï¼‰ï¼Œç›´æ¥ä½¿ç”¨info()ï¼ŒæŸ¥çœ‹æ•°æ®çš„ç»Ÿè®¡ç‰¹æ€§ã€‚
titanic.info()
```

```python
# æœºå™¨å­¦ä¹ æœ‰ä¸€ä¸ªä¸å¤ªè¢«åˆå­¦è€…é‡è§†ï¼Œå¹¶ä¸”è€—æ—¶ï¼Œä½†æ˜¯ååˆ†é‡è¦çš„ä¸€ç¯ï¼Œç‰¹å¾çš„é€‰æ‹©ï¼Œè¿™ä¸ªéœ€è¦åŸºäºä¸€äº›èƒŒæ™¯çŸ¥è¯†ã€‚æ ¹æ®æˆ‘ä»¬å¯¹è¿™åœºäº‹æ•…çš„äº†è§£ï¼Œsex, age, pclassè¿™äº›éƒ½å¾ˆæœ‰å¯èƒ½æ˜¯å†³å®šå¹¸å…ä¸å¦çš„å…³é”®å› ç´ ã€‚
X = titanic[['pclass', 'age', 'sex']]
y = titanic['survived']

# å¯¹å½“å‰é€‰æ‹©çš„ç‰¹å¾è¿›è¡Œæ¢æŸ¥ã€‚
X.info()
```

```python
# å€Ÿç”±ä¸Šé¢çš„è¾“å‡ºï¼Œæˆ‘ä»¬è®¾è®¡å¦‚ä¸‹å‡ ä¸ªæ•°æ®å¤„ç†çš„ä»»åŠ¡ï¼š
# 1) ageè¿™ä¸ªæ•°æ®åˆ—ï¼Œåªæœ‰633ä¸ªï¼Œéœ€è¦è¡¥å®Œã€‚
# 2) sex ä¸ pclassä¸¤ä¸ªæ•°æ®åˆ—çš„å€¼éƒ½æ˜¯ç±»åˆ«å‹çš„ï¼Œéœ€è¦è½¬åŒ–ä¸ºæ•°å€¼ç‰¹å¾ï¼Œç”¨0/1ä»£æ›¿ã€‚

# é¦–å…ˆæˆ‘ä»¬è¡¥å……ageé‡Œçš„æ•°æ®ï¼Œä½¿ç”¨å¹³å‡æ•°æˆ–è€…ä¸­ä½æ•°éƒ½æ˜¯å¯¹æ¨¡å‹åç¦»é€ æˆæœ€å°å½±å“çš„ç­–ç•¥ã€‚
X['age'] = X['age'].fillna(X['age'].mean())
```

```python
# å¯¹è¡¥å®Œçš„æ•°æ®é‡æ–°æ¢æŸ¥ã€‚
X.info()
```

```python
# ç”±æ­¤å¾—çŸ¥ï¼Œageç‰¹å¾å¾—åˆ°äº†è¡¥å®Œã€‚

# æ•°æ®åˆ†å‰²ã€‚
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state = 33)
```

```python
# æˆ‘ä»¬ä½¿ç”¨scikit-learn.feature_extractionä¸­çš„ç‰¹å¾è½¬æ¢å™¨
from sklearn.feature_extraction import DictVectorizer
vec = DictVectorizer(sparse=False)

# è½¬æ¢ç‰¹å¾åï¼Œæˆ‘ä»¬å‘ç°å‡¡æ˜¯ç±»åˆ«å‹çš„ç‰¹å¾éƒ½å•ç‹¬å‰¥ç¦»å‡ºæ¥ï¼Œç‹¬æˆä¸€åˆ—ç‰¹å¾ï¼Œæ•°å€¼å‹çš„åˆ™ä¿æŒä¸å˜ã€‚
X_train = vec.fit_transform(X_train.to_dict(orient='record'))
print( vec.feature_names_)
```

```python
# åŒæ ·éœ€è¦å¯¹æµ‹è¯•æ•°æ®çš„ç‰¹å¾è¿›è¡Œè½¬æ¢ã€‚
X_test = vec.transform(X_test.to_dict(orient='record'))
```

```python
# ä»sklearn.treeä¸­å¯¼å…¥å†³ç­–æ ‘åˆ†ç±»å™¨ã€‚
from sklearn.tree import DecisionTreeClassifier
# ä½¿ç”¨é»˜è®¤é…ç½®åˆå§‹åŒ–å†³ç­–æ ‘åˆ†ç±»å™¨ã€‚
dtc = DecisionTreeClassifier()
# ä½¿ç”¨åˆ†å‰²åˆ°çš„è®­ç»ƒæ•°æ®è¿›è¡Œæ¨¡å‹å­¦ä¹ ã€‚
dtc.fit(X_train, y_train)
# ç”¨è®­ç»ƒå¥½çš„å†³ç­–æ ‘æ¨¡å‹å¯¹æµ‹è¯•ç‰¹å¾æ•°æ®è¿›è¡Œé¢„æµ‹ã€‚
y_predict = dtc.predict(X_test)
```

```python
# ä»sklearn.metricså¯¼å…¥classification_reportã€‚
from sklearn.metrics import classification_report
# è¾“å‡ºé¢„æµ‹å‡†ç¡®æ€§ã€‚
print (dtc.score(X_test, y_test))
# è¾“å‡ºæ›´åŠ è¯¦ç»†çš„åˆ†ç±»æ€§èƒ½ã€‚
print (classification_report(y_predict, y_test, target_names = ['died', 'survived']))
```
